```python
from transformers import pipeline

classifier = pipeline("zero-shot-image-classification")
```

# ***Interpreter Layer Starts***

This layer generalises the top prediction.

```python
from google.colab import files
from PIL import Image

uploaded = files.upload()
image_path= list(uploaded.keys())[0]
image= Image.open(image_path)
candidates_labels=["fruit","vegetable","baked","meat","dairy"]
labels= classifier(image , candidate_labels=candidates_labels)
category=labels[0]["label"]
print(category)
```

# ***Feature Extraction - Brightness***

```python
import numpy as np

img_array=np.array(image)

brightness=np.mean(img_array)
print("Brightness:", brightness)

if brightness > 150:
  print("Freshness = Fresh")
elif brightness > 100:
  print("Freshness = Okay")
else:
   print("Freshness = Avoid")
```

```python
avg_color=np.mean(img_array, axis=(0,1))
print("Average Color:", avg_color)

red, green , blue = avg_color

if category=="fruit":
  if red > 150 and green > 100:
    print("Improved Freshness = Fresh")
  elif red > 100:
    print("Improved Freshness = Okay")
  else:
    print("Improved Freshness = Avoid")
else:
  print("Not Defined")
```
